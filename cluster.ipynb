```python
import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift
from collections import Counter
```

```python
data = pd.read_csv('checkins.csv', sep='|')
```

```python
data.columns = [i.strip() for i in data.columns]
```

```python
data['longitude'] = data.loc[:, 'longitude'].str.strip()
```

```python
data = data[data['longitude'] != '']
```

```python
data
```

```python
koord = data[['latitude', 'longitude']][:100000]
```

```python
koord['longitude'] = pd.to_numeric(koord.longitude)
```

```python
koord.dtypes
```

```python
clustering = MeanShift(bandwidth=0.1, n_jobs=-1).fit(koord)
```

```python
clustering.labels_
```

```python
a = Counter(clustering.labels_)
```

```python
b = {}
for i in a.keys():
    if a[i] >= 15:
        b[i] = a[i]
```

```python
len(b)
```

```python
koord['class'] = clustering.labels_
```

```python
koord[koord['class'] == 1]
```

```python
centr = []
for i in b.keys():
    tem = koord[koord['class'] == i].mean()
    centr.append([i, tem[0], tem[1]])
```

```python
centr
```

```python
of = [[33.751277, -118.188740], [25.867736, -80.324116], [51.503016, -0.075479], 
      [52.378894, 4.885084], [39.366487, 117.036146], [-33.868457, 151.205134]]
```

```python
rast = []
for j in of:
    for i in centr:
        rast.append([((i[1] - j[0])**2 + (i[2]-j[1])**2) ** 0.5, i])
```
