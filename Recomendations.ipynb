```python
def rec(data, count):
    res = []
    for i in data:
        ls = []
        for j in range(min(len(i), count)):
            ls.append(i[j][0])
        res.append(ls)    
    return res    
```

```python
def prec_k(rec, done, k):
    res = []
    for i in range(len(done)):
        c = 0
        for j in rec[i]:
            if j in done[i]:
                c += 1
        res.append(c/k) 
    print(np.mean(res))
```

```python
def recall_k(rec, done, k):
    res = []
    for i in range(len(done)):
        c = 0
        for j in done[i]:
            if j in rec[i]:
                c += 1
        res.append(c/len(done[i])) 
    print(np.mean(res))
```

```python
recall_k(rec_tr_se1, tr_done, 1)
prec_k(rec_tr_se1, tr_done, 1)
recall_k(rec_tr_se5, tr_done, 5)
prec_k(rec_tr_se5, tr_done, 5)
```
